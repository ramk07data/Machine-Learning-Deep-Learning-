```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('TexasTurbine (2).csv')
```

```python
data
```

```python
df=data.dropna()
```

```python
df
```

```python
df=df.rename({'Time stamp':'Time_stamp','System power generated | (kW)':'System_power_generated',
              'Wind speed | (m/s)':'Wind_speed','Wind direction | (deg)':'Wind_direction','Pressure | (atm)':'Pressure',
              "Air temperature | ('C)":'Air_temperature'},axis=1)
```

```python
df.columns
```

```python
df.head()
```

```python
# Histogram
plt.hist(df['System_power_generated'],color='violet',edgecolor='black')
plt.xlabel('System Power Generation in kW')
plt.title('Wind Power Generation')
plt.show()
```

```python
plt.hist(df['Wind_speed'],bins=9,color='green',edgecolor='black')
plt.xlabel('Wind Speed in m/s')
plt.title('Wind Speed')
```

```python
plt.boxplot(df['Wind_direction'])
plt.show()
```

```python
sns.boxplot(df['Pressure'],color='g')
```

```python
sns.boxplot(df['Air_temperature'])
plt.show()
```

```python
df.hist(figsize=(15,55),layout=(15,4))
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
